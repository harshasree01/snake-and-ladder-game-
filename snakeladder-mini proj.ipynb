```
import random

# Snakes and Ladders positions
snakes = {
    16: 6,
    48: 30,
    62: 19,
    88: 24,
    95: 56,
    97: 78
}

ladders = {
    1: 38,
    4: 14,
    9: 31,
    21: 42,
    28: 84,
    51: 67,
    72: 91,
    80: 99
}

player_position = 0

print("ðŸŽ² Welcome to Snake and Ladder Game ðŸŽ²")
print("Reach position 100 to win\n")

while player_position < 100:
    input("Press Enter to roll the dice...")
    dice = random.randint(1, 6)
    print("Dice rolled:", dice)

    if player_position + dice <= 100:
        player_position += dice

    if player_position in snakes:
        print("ðŸ Snake bite!")
        player_position = snakes[player_position]

    elif player_position in ladders:
        print("ðŸªœ Ladder climb!")
        player_position = ladders[player_position]

    print("Current position:", player_position)
    print("-" * 30)

if player_position == 100:
    print("ðŸŽ‰ Congratulations! You won the game ðŸŽ‰")
```
